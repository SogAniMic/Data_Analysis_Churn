# Application de fonctions aux colones et aux lignes

Pour appliquer des fonctions à chaque colonne, utiliez apply() 

```python
import numpy as np
```

```python
data.apply(np.max)
```

```python
data[data["State"].apply(lambda state : state[0] == "W")].head()
```

```python
d = {"No" : False, "Yes":True}
data["International plan"] = data["International plan"].map(d)
data.head
```

```python
data = data.replace({"Voice mail plan" : d})
data.head()
```

# Regroupement

Voici un exemple où nous regroupons les données en fonction des valeurs de la variable Churn et affichons les statistiques de 3 colonnes dans chaque groupe


```python
colonnes_a_afficher = ["Total day minutes", "Total eve minutes", "Total night minutes"]
data.groupby(["Churn"])[colonnes_a_afficher].mean()
```

```python
data.groupby(["Churn"])[colonnes_a_afficher].describe()
```

```python
data.groupby(["Churn"])[colonnes_a_afficher].agg([np.mean,np.std,np.min,np.max])
```
